### Условный оператор

В коде часто приходится проверять выполнимость или невыполнимость каких-то условий. Синтаксис:
<br>if <условие1 (булевское выражение)> :

<br><b><i>То есть, отступы позволяют понять, где начинается код, который должен выполняться при выполнении условия в if, и где заканчивается.</b></i>

Задание:
<br>Для функции $$sin(xe^{cos(x)})$$ на интервале [0;10] с шагом 0,05 построить график функции, закрасить площадь под ним и вычислить значение этой площади.
<br>Для вычисления площади воспользуйте функции trapz(), вычисляющей интеграл по правилу трапеции. Для ее корректной работы необходимо подключить следующие библиотеки:

## Задание 4.3.2

Дана функция $$ \frac{\sqrt{1 + e^{\sqrt{x}} + cosx^2}}{|1 - sin^3x|} + ln|2x|. $$ Создать массив из 10 значений функции (𝑥, например, изменяется от 1 до 10). Выделить срез первой половины массива и построить графики для основного массива – линейный и для среза –  точечный.

```python
import math as m
from matplotlib import pyplot as plt

def f(x):
    numerator = m.sqrt( 1 + m.exp(1) ** m.sqrt(x) + m.cos(x ** 2))
    denominator = abs( 1 - m.sin(x) ** 3)
    return numerator / denominator + m.log( abs(2 * x), m.exp(1) )

a = [ f(x) for x in range(1, 11) ]

# Установка размера окна
plt.figure(figsize=(5, 5))

# Создание графика
plt.plot(a)

# Установка заголовка для диаграммы
plt.title('Линейный график')

# Добавление сетки на диаграмму
plt.grid()

# Отображение диаграммы
plt.show()

# Установка размера окна
plt.figure(figsize=(5, 5))

# Установка заголовка для диаграммы
plt.title("Точечный график")

# Добавление сетки на диаграмму
plt.grid()

# Создание графика
plt.scatter(range(len(a[:5:])), a[:5:])

# Отображение диаграммы
plt.show()
```

## Задание 4.3.3

Дана функция $$ |cos(xe^{(cos(x)+ln(x+1))})|. $$ Построить график на интервале (0,10) с шагом 1 с заливкой площади и найти эту площадь под ним. Для вычисления площади воспользуйте функции trapz(), вычисляющей интеграл по правилу трапеции.

```python
import math as m
from matplotlib import pyplot as plt
from scipy.integrate import simps
from numpy import trapz

def f(x):
    pow = m.cos(x) + m.log( x + 1, m.exp(1) )
    return abs( m.cos( x * m.exp(1) ** pow ) )

a = [ f(x) for x in range(0, 10, 1) ]

# Установка размера окна
plt.figure(figsize=(5, 5))

# Создание графика, c - color
plt.plot([i for i in range(0, 10)], a, c = 'black')

# Закрашивание области под графиком
plt.fill_between([i for i in range(0, 10)], a)

# Установка заголовка для диаграммы
plt.title('Линейный график')

# Добавление сетки на диаграмму
plt.grid()

# Отображение диаграммы
plt.show()

# Вычисление площади
area = trapz(a)

print("Площадь под графиком: " + str(round(area, 3)))
```

## Задание 4.3.5

Создайте простейший калькулятор, включающий основные действия
для двух переменных ′ + ′, ′ − ′, ′ ∙ ′, ′ ÷ ′, а также вычисление
следующих функций: e ** (x + y), sin(x + y), cos(x + y), x ** y

```python
import math as m

def calculator():
  while True:
    num1 = float(input("Введите первое число: "))
    num2 = float(input("Введите второе число: "))
    operation = input("Введите операцию (+, -, *, /, exp, sin, cos, ^): ")
    
    if operation == '+':
      result = num1 + num2
    elif operation == '-':
      result = num1 - num2
    elif operation == '*':
      result = num1 * num2
    elif operation == '/':
      result = num1 / num2
    elif operation == 'exp':
        result = m.exp(1) ** (num1 + num2)
    elif operation == 'sin':
        result = m.sin(num1 + num2)
    elif operation == 'cos':
        result = m.cos(num1 + num2)
    elif operation == '^':
        result = num1 ** num2
    else:
      print("Некорректная операция!")
      continue
    
    print("Результат: ", result)
    
    again = input("Желаете продолжить вычисления (y/n)? ")
    if again.lower() != 'y':
      break
      
calculator()
```

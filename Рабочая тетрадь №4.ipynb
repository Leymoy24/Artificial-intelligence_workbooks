<br>–î–ª—è —Ä–µ—à–µ–Ω–∏—è –¥–∞–Ω–Ω–æ–π –∑–∞–¥–∞—á–∏ –≤ Python –µ—Å—Ç—å –º–Ω–æ–∂–µ—Å—Ç–≤–æ –±–∏–±–ª–∏–æ—Ç–µ–∫. –°–∞–º—ã–º–∏

<br><b>scipy - scipy.linalg</b> (—Å–æ–¥–µ—Ä–∂–∏—Ç –≤—Å–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∏–∑ numpy.linalg –ø–ª—é—Å —á–∞—Å—Ç—å
–Ω–æ–≤—ã—Ö —Ñ—É–Ω–∫—Ü–∏–π, –∫–æ—Ç–æ—Ä—ã—Ö –Ω–µ—Ç –≤ numpy.linalg).

### –ó–∞–¥–∞–Ω–∏–µ:
–ü—Ä–µ–¥—Å—Ç–∞–≤—å—Ç–µ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –∏ –ø–æ—Å—Ç—Ä–æ–π—Ç–µ —ç–∫—Ç—Ä–∞–ø–æ–ª—è—Ü–∏—é –ø–æ–ª–∏–Ω–æ–º–∞–º–∏
–ø–µ—Ä–≤–æ–π, –≤—Ç–æ—Ä–æ–π –∏ —Ç—Ä–µ—Ç—å–µ–π —Å—Ç–µ–ø–µ–Ω–∏.

```python
import numpy as np
import matplotlib.pyplot as plt

# –°–≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º —Å–ª—É—á–∞–π–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
x = np.linspace(0, 10, 50)
y = np.sin(x) + np.random.normal(0, 0.1, 50)

# –ü–æ—Å—Ç—Ä–æ–∏–º –ø–æ–ª–∏–Ω–æ–º—ã –ø–µ—Ä–≤–æ–π, –≤—Ç–æ—Ä–æ–π –∏ —Ç—Ä–µ—Ç—å–µ–π —Å—Ç–µ–ø–µ–Ω–∏
p1 = np.poly1d(np.polyfit(x, y, 1))
p2 = np.poly1d(np.polyfit(x, y, 2))
p3 = np.poly1d(np.polyfit(x, y, 3))

# –ù–∞—Ä–∏—Å—É–µ–º –≥—Ä–∞—Ñ–∏–∫–∏
plt.scatter(x, y)
plt.plot(x, p1(x), label="–ü–µ—Ä–≤–∞—è —Å—Ç–µ–ø–µ–Ω—å")
plt.plot(x, p2(x), label="–í—Ç–æ—Ä–∞—è —Å—Ç–µ–ø–µ–Ω—å")
plt.plot(x, p3(x), label="–¢—Ä–µ—Ç—å—è —Å—Ç–µ–ø–µ–Ω—å")
plt.legend()
plt.show()
```

3) ùëì(ùë•, ùëè) = ùëè<sub>0</sub> + ùëè<sub>1</sub>ùëôùëõ(ùë•)
4) ùëì(ùë•, ùëè) = ùëè<sub>0</sub> ùë•<sup>ùëè<sub>1</sub></sup>

–î–ª—è –ø–æ–¥—Å—Ç–∞–Ω–æ–≤–∫–∏ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –º–æ–∂–Ω–æ –∑–∞–º–µ–Ω–∏—Ç—å –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é xdata –Ω–∞ —Å–≤–æ–π –º–∞—Å—Å–∏–≤ —Ç–æ—á–µ–∫, –∞ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é ydata - –Ω–∞ —Å–≤–æ–∏ –∑–Ω–∞—á–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ f(x, b) —Å –Ω–µ–∫–æ—Ç–æ—Ä—ã–º –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ–º —à—É–º–∞.
<br>–î–ª—è –∞–Ω–∞–ª–∏–∑–∞ –¥–∏–Ω–∞–º–∏–∫–∏ –∏–∑–º–µ–Ω–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –º–æ–∂–Ω–æ –ø–æ—ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å —Å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–º —à—É–º–∞, –∫–æ—Ç–æ—Ä—ã–π –¥–æ–±–∞–≤–ª—è–µ—Ç—Å—è –∫ –¥–∞–Ω–Ω—ã–º. –ù–∞–ø—Ä–∏–º–µ—Ä, –º–æ–∂–Ω–æ —É–≤–µ–ª–∏—á–∏—Ç—å –∏–ª–∏ —É–º–µ–Ω—å—à–∏—Ç—å –µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å, –∫–∞–∫ —ç—Ç–æ –ø–æ–≤–ª–∏—è–µ—Ç –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —ç–∫—Å—Ç—Ä–∞–ø–æ–ª—è—Ü–∏–∏.
<br>–¢–∞–∫–∂–µ –º–æ–∂–Ω–æ –ø–æ–∏–≥—Ä–∞—Ç—å —Å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º —Ç–æ—á–µ–∫ –≤ –º–∞—Å—Å–∏–≤–µ xdata. –ù–∞–ø—Ä–∏–º–µ—Ä, —É–≤–µ–ª–∏—á–∏—Ç—å –µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏–µ, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å –±–æ–ª–µ–µ –≥–ª–∞–¥–∫–∏–π –≥—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏, –∏–ª–∏ —É–º–µ–Ω—å—à–∏—Ç—å, —á—Ç–æ–±—ã –ø–æ–ª—É—á–∏—Ç—å –±–æ–ª–µ–µ —à—É–º–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫.

```python
import scipy as sp
from scipy.optimize import curve_fit

# f(x, b) = b0 + b1x
# –¥–æ–±–∞–≤–∏–º —à—É–º–∞ –≤ –¥–∞–Ω–Ω—ã–µ, —Å–¥–µ–ª–∞–Ω–Ω—ã–µ –ø–æ —Ñ—É–Ω–∫—Ü–∏–∏ f(x, b) —Å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º–∏ b = (0.25, 0.75)
beta = (0.25, 0.75)
def f(x, b0, b1):
    return b0 + b1 * x

# –∑–∞–¥–∞–¥–∏–º –º–∞—Å—Å–∏–≤ —Ç–æ—á–µ–∫ xi
xdata = np.linspace(0, 5, 50)

# —Å–æ–∑–¥–∞–µ–º —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏ –ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ç–æ—á–µ–∫ yi (–±–µ–∑ —à—É–º–∞)
y = f(xdata, *beta)

# –∑–∞—à—É–º–ª—è–µ–º —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ
ydata = y + 0.05 * np.random.randn(len(xdata))

# –∏—Å–ø–æ–ª—å–∑—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Ä–µ—à–µ–Ω–∏—è –≤ –≤–∏–¥–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ñ–∞–Ω–∫—Ü–∏–∏ f(x) –¥–ª—è —É–∫–∞–∑–∞–Ω–Ω—ã—Ö xdata –∏ ydata
beta_opt, beta_cov = sp.optimize.curve_fit(f, xdata, ydata)
print(beta_opt)

# –≤—ã—á–∏—Å–ª–∏–º –ª–∏–Ω–µ–π–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
lin_dev = sum(beta_cov[0])
print(lin_dev)

# –≤—ã—á–∏—Å–ª–∏–º –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
residuals = ydata - f(xdata, *beta_opt)
fres = sum(residuals**2)
print(fres)

fig, ax = plt.subplots()
ax.scatter(xdata, ydata)
ax.plot(xdata, y, 'r', lw=2)
ax.plot(xdata, f(xdata, *beta_opt), 'b', lw=2)
ax.set_xlim(0, 5)
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$f(x, \beta)$", fontsize=18)
plt.show()
```

```python
import scipy as sp
from scipy.optimize import curve_fit

# ùëì(ùë•, ùëè) = ùëè0 + ùëè1ùë• + ùëè2ùë•^2
# –¥–æ–±–∞–≤–∏–º —à—É–º–∞ –≤ –¥–∞–Ω–Ω—ã–µ, —Å–¥–µ–ª–∞–Ω–Ω—ã–µ –ø–æ —Ñ—É–Ω–∫—Ü–∏–∏ f(x, b) —Å –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞–º–∏ b = (0.25, 0.75, 0.5)
beta = (0.25, 0.75, 0.5)
def f(x, b0, b1, b2):
    return b0 + b1 * x + b2 * x * x

# –∑–∞–¥–∞–¥–∏–º –º–∞—Å—Å–∏–≤ —Ç–æ—á–µ–∫ xi
xdata = np.linspace(0, 5, 50)

# —Å–æ–∑–¥–∞–µ–º —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏ –ø—Ä–∞–≤–∏–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ç–æ—á–µ–∫ yi (–±–µ–∑ —à—É–º–∞)
y = f(xdata, *beta)

# –∑–∞—à—É–º–ª—è–µ–º —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ
ydata = y + 0.05 * np.random.randn(len(xdata))

# –∏—Å–ø–æ–ª—å–∑—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Ä–µ—à–µ–Ω–∏—è –≤ –≤–∏–¥–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤ —Ñ–∞–Ω–∫—Ü–∏–∏ f(x) –¥–ª—è —É–∫–∞–∑–∞–Ω–Ω—ã—Ö xdata –∏ ydata
beta_opt, beta_cov = sp.optimize.curve_fit(f, xdata, ydata)
print(beta_opt)

# –≤—ã—á–∏—Å–ª–∏–º –ª–∏–Ω–µ–π–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
lin_dev = sum(beta_cov[0])
print(lin_dev)

# –≤—ã—á–∏—Å–ª–∏–º –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
residuals = ydata - f(xdata, *beta_opt)
fres = sum(residuals**2)
print(fres)

fig, ax = plt.subplots()
ax.scatter(xdata, ydata)
ax.plot(xdata, y, 'r', lw=2)
ax.plot(xdata, f(xdata, *beta_opt), 'b', lw=2)
ax.set_xlim(0, 5)
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$f(x, \beta)$", fontsize=18)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# –∑–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ –∏–∑ csv-—Ñ–∞–π–ª–∞
data = pd.read_csv('https://raw.githubusercontent.com/AnnaShestova/salary-years-simple-linear-regression/master/Salary_Data.csv')

# —Ä–∞–∑–¥–µ–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –∞—Ç—Ä–∏–±—É—Ç—ã –∏ –º–µ—Ç–∫–∏
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

# —Ç–µ–ø–µ—Ä—å, –∫–æ–≥–¥–∞ —É –Ω–∞—Å –µ—Å—Ç—å –∞—Ç—Ä–∏–±—É—Ç—ã –∏ –º–µ—Ç–∫–∏, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Ä–∞–∑–¥–µ–ª–∏—Ç—å –∏—Ö –Ω–∞ –æ–±—É—á–∞—é—â–∏–π –∏ —Ç–µ—Å—Ç–æ–≤—ã–π –Ω–∞–±–æ—Ä—ã.
# –ø—Ä–∏–≤–µ–¥–µ–Ω–Ω—ã–π —Ñ—Ä–∞–≥–º–µ–Ω—Ç —Ä–∞–∑–¥–µ–ª—è–µ—Ç 80% –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á–∞—é—â–∏–π –Ω–∞–±–æ—Ä, –∞ 20% –¥–∞–Ω–Ω—ã—Ö - –Ω–∞ –Ω–∞–±–æ—Ä —Ç–µ—Å—Ç–æ–≤
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# —Å–æ–∑–¥–∞–µ–º –æ–±—ä–µ–∫—Ç –º–æ–¥–µ–ª–∏ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
regressor = LinearRegression()

# –æ–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å –Ω–∞ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
regressor.fit(X_train, y_train)

# –ø–æ–ª—É—á–∞–µ–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ª–∏–Ω–∏–∏ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
b0 = regressor.intercept_
b1 = regressor.coef_[0]

# –≤—ã–≤–æ–¥–∏–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã
print('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –ª–∏–Ω–∏–∏ —Ä–µ–≥—Ä–µ—Å—Å–∏–∏: b0 = {:.2f}, b1 = {:.2f}'.format(b0, b1))

# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –¥–∞–Ω–Ω—ã—Ö –∏ –ª–∏–Ω–∏—é —Ä–µ–≥—Ä–µ—Å—Å–∏ –¥–ª—è –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏
plt.scatter(X, y, color='blue')
plt.plot(X, regressor.predict(X), color='red')
plt.xlabel('–û–ø—ã—Ç —Ä–∞–±–æ—Ç—ã, –ª–µ—Ç')
plt.ylabel('–ó–∞—Ä–∞–±–æ—Ç–Ω–∞—è –ø–ª–∞—Ç–∞, —Ä—É–±.')
plt.show()

# —Å—Ç—Ä–æ–∏–º –ø—Ä–æ–≥–Ω–æ–∑ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
y_pred = regressor.predict(X_test)

# —Å—Ä–∞–≤–Ω–∏–º —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º—ã–º–∏
df = pd.DataFrame({'–§–∞–∫—Ç–∏—á–µ—Å–∫–∏–π' : y_test, '–ü—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º—ã–π' : y_pred})
print(df)

# —Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –¥–∞–Ω–Ω—ã—Ö –∏ –ª–∏–Ω–∏—é —Ä–µ–≥—Ä–µ—Å—Å–∏ –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red')
plt.xlabel('–û–ø—ã—Ç —Ä–∞–±–æ—Ç—ã, –ª–µ—Ç')
plt.ylabel('–ó–∞—Ä–∞–±–æ—Ç–Ω–∞—è –ø–ª–∞—Ç–∞, —Ä—É–±.')
plt.show

# –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –≤ –≤–∏–¥–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã
df.plot(kind='bar')
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()

# –¥–µ–ª–∞–µ–º –ø—Ä–æ–≥–Ω–æ–∑ –¥–ª—è –Ω–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö
new_X = [[10], [15]]
new_y = regressor.predict(new_X)
print('–ü—Ä–æ–≥–Ω–æ–∑ –∑–∞—Ä–∞–±–æ—Ç–Ω–æ–π –ø–ª–∞—Ç—ã –¥–ª—è –æ–ø—ã—Ç–∞ —Ä–∞–±–æ—Ç—ã –≤ 10 –ª–µ—Ç:', round(new_y[0], 2))
print('–ü—Ä–æ–≥–Ω–æ–∑ –∑–∞—Ä–∞–±–æ—Ç–Ω–æ–π –ø–ª–∞—Ç—ã –¥–ª—è –æ–ø—ã—Ç–∞ —Ä–∞–±–æ—Ç—ã –≤ 15 –ª–µ—Ç:', round(new_y[1], 2))
```

### –ó–∞–¥–∞–Ω–∏–µ
–ü–æ—Å—Ç—Ä–æ–π—Ç–µ –º–æ–¥–µ–ª—å –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –¥–ª—è –ø—Ä–æ–∏–∑–≤–æ–ª—å–Ω—ã—Ö
–¥–∞–Ω–Ω—ã—Ö –∏–∑ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö —Å—Ç–æ–ª–±—Ü–æ–≤. –î–ª—è –ø—Ä–∏–º–µ—Ä–∞ –º–æ–∂–Ω–æ –≤–∑—è—Ç—å –ø–æ—Ç—Ä–µ–±–ª–µ–Ω–∏—è
–≥–∞–∑–∞ (–≤ –º–∏–ª–ª–∏–æ–Ω–∞—Ö –≥–∞–ª–ª–æ–Ω–æ–≤) –≤ 48 —à—Ç–∞—Ç–∞—Ö –°–®–ê –∏–ª–∏ –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö –æ
–∫–∞—á–µ—Å—Ç–≤–µ –∫—Ä–∞—Å–Ω–æ–≥–æ –≤–∏–Ω–∞ (1) –∏ (2) —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ. –ù–∞–π–¥–∏—Ç–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã
–º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –ø—Ä–æ–≥–Ω–æ–∑.

```python
import pandas as pd
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics

# –∑–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ –∏–∑ csv-—Ñ–∞–π–ª–∞
data = pd.read_csv('https://raw.githubusercontent.com/likarajo/petrol_consumption/master/data/petrol_consumption.csv')
data = data.rename(columns={'Petrol_tax': '–ù–∞–ª–æ–≥ –Ω–∞ –±–µ–Ω–∑–∏–Ω', 
                            'Average_income': '–°—Ä–µ–¥–Ω–∏–π –¥–æ—Ö–æ–¥', 
                            'Paved_Highways': '–ê—Å—Ñ–∞–ª—å—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –¥–æ—Ä–æ–≥–∏',
                            'Population_Driver_licence(%)': '–ù–∞—Å–µ–ª–µ–Ω–∏–µ —Å –≤–æ–¥–∏—Ç–µ–ª—å—Å–∫–∏–º–∏ –ø—Ä–∞–≤–∞–º–∏ (%)',
                            'Petrol_Consumption': '–ü–æ—Ç—Ä–µ–±–ª–µ–Ω–∏–µ –±–µ–Ω–∑–∏–Ω–∞'})
print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:")
print(data.head(), end="\n\n")

# —Ä–∞–∑–¥–µ–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –∞—Ç—Ä—Ä–∏–±—É—Ç—ã (–ø–µ—Ä–≤—ã–µ 4 —Å—Ç–æ–±—Ü–∞) –∏ –º–µ—Ç–∫–∏ (–ø–æ—Å–ª–µ–¥–Ω–∏–π —Å—Ç–æ–ª–±–µ—Ü)
X = data.drop('–ü–æ—Ç—Ä–µ–±–ª–µ–Ω–∏–µ –±–µ–Ω–∑–∏–Ω–∞', axis=1)
y = data['–ü–æ—Ç—Ä–µ–±–ª–µ–Ω–∏–µ –±–µ–Ω–∑–∏–Ω–∞']

# —Ä–∞–∑–¥–µ–ª—è–µ–º –¥–∞–Ω–Ω—ã–µ –Ω–∞ –æ–±—É—á–∞—é—â–∏–π –∏ —Ç–µ—Å—Ç–æ–≤—ã–π –Ω–∞–±–æ—Ä—ã
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# —Å—Ç—Ä–æ–∏–º –º–æ–¥–µ–ª—å –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# –≤—ã–≤–æ–¥–∏–º –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã –º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
coeff_df = pd.DataFrame(regressor.coef_, X.columns, columns=['–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç'])
print(coeff_df, end="\n\n")

# –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –ø–æ—Ç—Ä–µ–±–ª–µ–Ω–∏—è —Ç–æ–ø–ª–∏–≤–∞
y_pred = regressor.predict(X_test)
df = pd.DataFrame({'–§–∞–∫—Ç–∏—á–µ—Å–∫–∏–π' : y_test, '–ü—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º—ã–π' : y_pred})
print(df, end="\n\n")

print("–°—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞:", metrics.mean_squared_error(y_test, y_pred))
```

### –ó–∞–¥–∞–Ω–∏–µ*
–ó–∞–¥–∞—á–∞: –≠–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞–ª—å–Ω–æ –ø–æ–ª—É—á–µ–Ω—ã N ‚àí –∑–Ω–∞—á–µ–Ω–∏–π –≤–µ–ª–∏—á–∏–Ω—ã Y –ø—Ä–∏
—Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏—è—Ö –≤–µ–ª–∏—á–∏–Ω—ã X. –ü–æ—Å—Ç—Ä–æ–∏—Ç—å –ø–æ–ª–∏–Ω–æ–º—ã –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π
—Å—Ç–µ–ø–µ–Ω–∏, –∞–ø–ø—Ä–æ–∫—Å–∏–º–∏—Ä—É—é—â–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞, —Å –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ–º
–º–µ—Ç–æ–¥–∞ –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤. –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤—ã–≤–æ–¥—è—Ç—Å—è –≤ –≤–∏–¥–µ —Ç–∞–±–ª–∏—Ü
–∑–Ω–∞—á–µ–Ω–∏–π –∏ –≥—Ä–∞—Ñ–∏–∫–æ–≤, –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –ø–æ–ª–∏–Ω–æ–º–æ–≤.

| 3,0 | 2,0 |


```python
import numpy as np
import matplotlib.pyplot as plt

# –í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
X = np.array([2.0, 2.2, 2.4, 2.6, 2.8, 3.0])
Y = np.array([4.0, 2.0, 4.0, 2.0, 5.0, 2.0])

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
plt.scatter(X, Y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# –ê–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏—è –ø–æ–ª–∏–Ω–æ–º–æ–º –ø–µ—Ä–≤–æ–π —Å—Ç–µ–ø–µ–Ω–∏
p1 = np.polyfit(X, Y, 1)
print("–£—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø–æ–ª–∏–Ω–æ–º–∞ –ø–µ—Ä–≤–æ–π —Å—Ç–µ–ø–µ–Ω–∏: y = {:.2f}x + {:.2f}".format(p1[0], p1[1]))

# –ê–ø–ø—Ä–æ–∫—Å–∏–º–∞—Ü–∏—è –ø–æ–ª–∏–Ω–æ–º–æ–º –≤—Ç–æ—Ä–æ–π —Å—Ç–µ–ø–µ–Ω–∏
p2 = np.polyfit(X, Y, 2)
print("–£—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø–æ–ª–∏–Ω–æ–º–∞ –≤—Ç–æ—Ä–æ–π —Å—Ç–µ–ø–µ–Ω–∏: y = {:.2f}x^2 + {:.2f}x + {:.2f}".format(p2[0], p2[1], p2[2]))

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
plt.scatter(X, Y)
plt.plot(X, np.polyval(p1, X), label='–ü–æ–ª–∏–Ω–æ–º –ø–µ—Ä–≤–æ–π —Å—Ç–µ–ø–µ–Ω–∏')
plt.plot(X, np.polyval(p2, X), label='–ü–æ–ª–∏–Ω–æ–º –≤—Ç–æ—Ä–æ–π —Å—Ç–µ–ø–µ–Ω–∏')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# –í—ã–≤–æ–¥ —Ç–∞–±–ª–∏—Ü—ã –∑–Ω–∞—á–µ–Ω–∏–π
print("x\t\ty\t\ty (–ø–æ–ª–∏–Ω–æ–º 1)\ty (–ø–æ–ª–∏–Ω–æ–º 2)")
for i in range(len(X)):
    F1 = np.polyval(p1, X[i])
    F2 = np.polyval(p2, X[i])
    print("{:.2f}\t\t{:.2f}\t\t{:.2f}\t\t{:.2f}".format(X[i], Y[i], F1, F2))
```
